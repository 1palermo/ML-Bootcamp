```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#generate data using a dumy function
X = np.arange(10)
Y = (X-5)**2+3

print(X,Y)
```

```python
#visualzing the given data
plt.style.use("seaborn")
plt.xlabel("X")
plt.ylabel("y")
plt.plot(X,Y)
plt.show()
```

```python
#gradient Descent
#intializing X
x  = 0
lr = 0.1

plt.plot(X,Y)
for i in range(50):
    grad = 2*(x-5)
    x = x - lr * grad
    y = (x-5)**2+3
    plt.scatter(x,y)
    print(x)
plt.show()
```
